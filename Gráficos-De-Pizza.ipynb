```python
import matplotlib.pyplot as plt
```

```python
# série de Dados....
marcas = ['Ford', 'Chevrolet', 'Toyota', 'Hyundai', 'Volksagen']
vendas = [350, 400, 300, 250, 200]
```

```python
explode = [0, 0.1, 0, 0, 0]

cores = ['Skyblue', 'lightgreen', 'coral', 'gold', 'lightpink']

fig = plt.figure(figsize=(6,4)) # Ajustar o tamanho da figura....

fatias_props = {'Edgecolor' : 'green',
                'linewidth' : 1.5,
                'linestyle' : 'dotted',
                'alpha' : 0.6
}

plt.pie(
    vendas,
    labels = marcas,
    autopct = "%1.1f%%",
    explode = explode,
    shadow = True,
    colors = cores,
    wedgeprops = fatias_props
)

plt.legend(marcas, title = "Marcas", loc = "upper right", bbox_to_anchor=(1.7, 1))
plt.title("Vendas Anuais de Veículos por Marcas.")
plt.show()
```
